## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)
- A geodatabase dataset helps organize map data inside a geodatabase like a folder. Feature classes would be what goes inside that folder. Feature classes contain the map information like points for locations, shapes for areas and lines for roads. Feature datasets keep related feature classes together and keep them on the same coordinate system so everything lines up. This keeps things organized to make mapping less confusing. The relationship between them is like a Russian nesting doll. Each layer gets smaller adding a new level of organization and detail. The structure of this relationship is key and really simplifies ArcGIS once you understand it.
